```python
import pandas as pd
import plotly as py
import plotly.graph_objs as go

# command for work offline
py.offline.init_notebook_mode(connected=True)
```

```python
prices = pd.read_csv('prices.csv')
```

```python
prices
```

```python
prices['type'][1] = 'Diagnosis'
```

```python
grp = ['type', 'country']
cost_type = prices.groupby(grp).min()
```

```python
prices
```

```python
cost_type.reset_index(inplace=True)
```

```python
cost_type
```

```python
x = ['Diagnosis', 'Drug', 'Surgery']
y1 = cost_type[cost_type['country'] == 'Spain']['value']
y2 = cost_type[cost_type['country'] == 'UK']['value']
y3 = cost_type[cost_type['country'] == 'Switzerland']['value']
y4 = cost_type[cost_type['country'] == 'US']['value']
trace1 = go.Bar(
    x=x,
    y=y1,
    name='Spain'
    
)
trace2 = go.Bar(
    x=x,
    y=y2,
    name='UK'
)

trace3 = go.Bar(
    x=x,
    y=y3,
    name='Switzerland'
)

trace4 = go.Bar(
    x=x,
    y=y4,
    name='USA'
)

data = [trace1, trace2, trace3, trace4]
layout = go.Layout(
    barmode='group'
)

fig = go.Figure(data=data, layout=layout)
```

```python
py.offline.iplot(fig)
```

```python
grp = ['country', 'type']
cost_type1 = prices.groupby(grp).min()
```

```python
cost_type1
```

```python
cost_type1.reset_index(inplace=True)
```

```python
cost_type1
```

```python
y = ['Spain', 'Switzerland', 'UK', 'USA']
x1 = cost_type1[cost_type1['type'] == 'Diagnosis']['value']
x2 = cost_type1[cost_type1['type'] == 'Drug']['value']
x3 = cost_type1[cost_type1['type'] == 'Surgery']['value']

trace1 = go.Bar(
    x=x1,
    y=y,
    name = 'Diagnosis',
    orientation = 'h',
    #text = cost_type1[cost_type1['type'] == 'Diagnosis']['intervention'],
    marker=dict(
        color=['grey', 'grey',
               'grey', 'rgba(222,45,38,0.8)'],
        line=dict(
            color='rgb(8,48,107)',
            width=1.5))
    
)
trace2 = go.Bar(
    x=x2,
    y=y,
    name='Drug',
    #text = cost_type1[cost_type1['type'] == 'Drug']['intervention'],
    textposition = 'auto',
    orientation = 'h',
        marker=dict(
        color=['grey', 'grey',
               'grey', 'rgba(222,45,38,0.8)'],
        line=dict(
            color='rgb(8,48,107)',
            width=1.5)),
    
)

trace3 = go.Bar(
    x=x3,
    y=y,
    name='Surgery',
    #text = cost_type1[cost_type1['type'] == 'Surgery']['intervention'],
    textposition = 'auto',
    orientation = 'h',
        marker=dict(
        color=['grey', 'grey',
               'grey', 'rgba(222,45,38,0.8)'],
        line=dict(
            color='rgb(8,48,107)',
            width=1.5)),
    
)

updatemenus = list([
    dict(active=-1,
         buttons=list([   
            dict(label = 'Diagnosis',
                 method = 'update',
                 args = [{'visible': [True, False, False]},
                         {'title': 'Diagnosis'}]),
            dict(label = 'Drug',
                 method = 'update',
                 args = [{'visible': [False, True, False]},
                         {'title': 'Drug',}]),
            dict(label = 'Surgery',
                 method = 'update',
                 args = [{'visible': [False, False, True]},
                         {'title': 'Surgery',}])
        ]),
    )
])
annotations = list([
    dict(text='Cost For:', x=0.2, y=1.085, yref='paper', align='left', showarrow=True)
])


data = [trace1, trace2, trace3]
layout = go.Layout(
    barmode='group',
    updatemenus=updatemenus,
    bargroupgap=0.1,
    showlegend = False,
    title='Comparison of average Spendings on Medication',
    xaxis=dict(
        title = 'Average Cost (in US dollars)',
        tickfont=dict(
            size=14,
            color='rgb(107, 107, 107)'
        )
    ),
    yaxis=dict(
        title='Countries',
        titlefont=dict(
            size=16,
            color='rgb(107, 107, 107)'
        ),
        tickfont=dict(
            size=14,
            color='rgb(107, 107, 107)'
        )
    )
)

fig = go.Figure(data=data, layout=layout)
```

```python
py.offline.iplot(fig)
```
