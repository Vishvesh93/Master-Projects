```python
import pandas as pd
import numpy as np
import plotly as plt

# command for work offline
plt.offline.init_notebook_mode(connected=True)
```

```python
map = pd.read_csv("who.csv")
map
```

```python
map.dropna(inplace = True)
```

```python
#df = map.pivot_table(values='Adult Mortality', columns =[map['Country'], map['Adult Mortality']], aggfunc = 'mean')
```

```python
map = map[map['Infant_Deaths']<100]
```

```python

data = [ dict(
        type = 'choropleth',
        locations = map['Code'],
        z = map['Infant_Deaths'],
        text = map['Country'],
        colorscale = [[0,"rgb(5, 10, 172)"],[0.35,"rgb(40, 60, 190)"],[0.5,"rgb(70, 100, 245)"],\
            [0.6,"rgb(90, 120, 245)"],[0.7,"rgb(106, 137, 247)"],[1,"rgb(220, 220, 220)"]],
        autocolorscale = False,
        reversescale = True,
        marker = dict(
            line = dict (
                color = 'rgb(180,180,180)',
                width = 0.5
            ) ),
        colorbar = dict(
            autotick = False,
            title = 'Average Infant Death per 1000 births'),
      ) ]

layout = dict(
    title = 'Comparison of average Infant deaths across all countries',
    geo = dict(
        showframe = False,
        showcoastlines = False,
        projection = dict(
            type = 'Mercator'
        )
    )
)

fig = dict( data=data, layout=layout )
plt.offline.iplot( fig, validate=False, filename='d3-world-map' )
```
